```python
df['Age'] = df['Rings'] + 1.5
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
ode=OrdinalEncoder(categories=[['I','M','F']])
df['Sex_ode']=ode.fit_transform(df[['Sex']])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['Rings','Age','Sex_ode','Sex'],axis=1)
y=df['Rings']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=20)
```

```python
xtrain
```

```python
ytrain
```

```python
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
mm_xtrain=mm.fit_transform(xtrain)
mm_xtest=mm.fit_transform(xtest)
```

```python
mm_xtrain
```

```python
# applying random forest classifier to predict age 
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
```

```python
# Fitting data into algorithm

fit_rfc=rfc.fit(mm_xtrain,ytrain)
rfc_score=rfc.score(mm_xtest,ytest)
```

```python
rfc_score
```
